```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = [6, 6]
```

```python
img = cv2.imread(root_path + "pepper.jpg")
plt.imshow(img[:,:,::-1])
plt.axis("off")
```

```python
def rgb_filter(img, kernel, mode="full", boundary="fill"):
    b, g, r = cv2.split(img)
    b = signal.correlate2d(b, kernel, mode=mode, boundary=boundary)
    g = signal.correlate2d(g, kernel, mode=mode, boundary=boundary)
    r = signal.correlate2d(r, kernel, mode=mode, boundary=boundary)
    output = cv2.merge([b, g, r])
    return output
```

```python
kernel = np.zeros((21, 21), dtype="uint8")
kernel[10, 10] = 1
fimage = rgb_filter(img, kernel, mode="full", boundary="fill")

print("Original shape: ", img.shape)
print("Filtered shape: ", fimage.shape)

plt.subplot(1, 2, 1)
plt.imshow(img[:,:,::-1])
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(fimage[:,:,::-1])
plt.axis("off")
```

```python
fimage = rgb_filter(img, kernel, mode='full', boundary='wrap')

#check the size of image
print("Original shape: ", img.shape)
print("Filtered shape: ", fimage.shape)

#view the image
plt.subplot(1, 2, 1); plt.imshow(img[:,:,::-1]); plt.axis("off")
plt.subplot(1, 2, 2); plt.imshow(fimage[:,:,::-1]); plt.axis("off")
```

## Unsharp masking

```python
blur = cv2.GaussianBlur(img, (7,7), 3.0)
fimage = cv2.addWeighted(img, 1.7, blur, -0.5, 0)
# fimage = image * 1.7 - 0.5 * blur + 0
plt.subplot(1, 3, 1); plt.imshow(img[:,:,::-1]); plt.axis("off"); plt.title("original")
plt.subplot(1, 3, 2); plt.imshow(blur[:,:,::-1]); plt.axis("off"); plt.title("blur (unsharp mask)")
plt.subplot(1, 3, 3); plt.imshow(fimage[:,:,::-1]); plt.axis("off"); plt.title("sharpen")
```
